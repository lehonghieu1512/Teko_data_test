```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/lehonghieu/Downloads/2014-new-york-city-taxi-trips/nyc_taxi_data_2014.csv')
```

```python
df.info()
```

As there are no missing values, we can carry on inspecting data

```python
df.describe()
```

```python
df['pickup_longitude'].hist()
df['pickup_latitude'].hist()
```

```python
df['dropoff_longitude'].hist()
df['dropoff_latitude'].hist()
```

We found that there are some outliners that will spoil our visualization. We will apply some limits on them to eliminate such outliners

```python
pickup_longitude_lim = [-74.177492, -73]
pickup_latitude_lim = [35, 40.982063]

dropoff_longitude_lim = [-74.177492, -73]
dropoff_latitude_lim = [35, 40.982063]
```

```python
df = df[(df.pickup_longitude>=pickup_longitude_lim[0]) & (df.pickup_longitude<=pickup_longitude_lim[1])]
df = df[(df.pickup_latitude>=pickup_latitude_lim[0]) & (df.pickup_latitude<=pickup_latitude_lim[1])]
df = df[(df.dropoff_longitude>=dropoff_longitude_lim[0]) & (df.dropoff_longitude<=dropoff_longitude_lim[1])]
df = df[(df.dropoff_latitude>=dropoff_latitude_lim[0]) & (df.dropoff_latitude<=dropoff_latitude_lim[1])]
```

```python
plt.plot(df['pickup_longitude'], df['pickup_latitude'], '.', color='b', alpha=0.2)
plt.show()
plt.plot(df['dropoff_longitude'], df['dropoff_latitude'], '.', color='r', alpha=0.1)
plt.show()
```

The 2 charts above show pickup and dropoff coordinates.

We can possibly predict the customers' places for traveling by looking for the pickup and dropoff points that overlap the most. 

```python
plt.plot(df['passenger_count'], df['trip_distance'], '.', color='b', alpha=0.2)
plt.show()
```

The chart above shows the connection between the number of passengers and trip distance. There is a really interesting fact that the lower the number of customer is the more likely they will travel great distances

```python
weekday_list = ['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['week_day'] = df['pickup_datetime'].dt.weekday_name

total_distance_by_weekday = df.groupby('week_day')['trip_distance'].sum()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
distance_by_day = [total_distance_by_weekday[date] for date in day_order]

plt.bar(day_order,distance_by_day)
plt.show()
```

The chart above shows the connection between weekdays and the distances they travel. And, of course, the closer it is to weekend the more likely they will travel further

```python
df['month'] = df['pickup_datetime'].dt.month_name()
total_distance_by_month= df.groupby('month')['trip_distance'].sum().to_dict()
plt.bar(total_distance_by_month.keys(),total_distance_by_month.values())
```

```python
# average price per mile

# eliminate 0 dist
df = df[df['trip_distance']>0]

df['price_per_miles'] = (df['total_amount'] - df['tip_amount']-df['tolls_amount'])/df['trip_distance']
```

```python
price_by_month = df.groupby('month')['price_per_miles'].mean().to_dict()
plt.plot(price_by_month.keys(), price_by_month.values(), '-')
```

```python
plt.plot(df['surcharge'], df['trip_distance'], '.')
plt.show()
plt.plot(df['surcharge'], df['passenger_count'], '.')
plt.show()
plt.plot(df['surcharge'], df['total_amount'], '.')
plt.show()
```

The charts above show the connections between surcharge and trip distance, passenger count, total amount respectively. We found that there is a fairly strong correlation between surcharge and trip distance/total amount

```python
plt.plot(df['tip_amount'], df['trip_distance'], '.')
plt.show()
plt.plot(df['tip_amount'], df['total_amount']-df['tip_amount'], '.')
plt.show()
plt.plot(df['tip_amount'], df['passenger_count'], '.')
plt.show()
```

The charts above show the connection between tip and trip distance, paid amount, passenger count respectively.
There is a tendency that the lower the number of passengers on that trip, the more likely they will tip the drivers more

```python
plt.figure(figsize = (16,16))
sns.heatmap(df.select_dtypes(exclude=['object', 'datetime64']).corr(), annot=True)
```

In overall, there is not much connection among all attributes except for amounts and trip_distance which is evident.
